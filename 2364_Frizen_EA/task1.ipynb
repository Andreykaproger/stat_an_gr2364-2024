## Основные теоретические положения

Многомерное нормальное распределение:

$ \overrightarrow{\xi} \sim \mathcal{N}(\overrightarrow{\mu}, \Sigma) $

$ p_{\overrightarrow{\xi}} \left(\overrightarrow{x}\right) = \cfrac{1}{\left(2\pi\right)^{n/2} \sqrt{\det\Sigma}} \cdot \exp \left(-\cfrac{1}{2}\left(\overrightarrow{x}-\mu\right)^{\intercal}\Sigma^{-1}\left(\overrightarrow{x}-\mu\right)\right) $

Коэффициент ковариации:

$ \rho (\xi, \eta) = \cfrac{\mathrm{cov} (\xi, \eta)}{\sqrt{\mathbb{D} \xi \cdot \mathbb{D} \eta}} $

Афинное преобразование:

$ \begin{pmatrix}
x\\
y
\end{pmatrix} = A \begin{pmatrix}
u\\
v
\end{pmatrix} + \begin{pmatrix}
\alpha\\
\beta
\end{pmatrix} $

Ортогональное преобразование:

$ \Sigma_{id} = Q \Sigma Q^{\intercal} $

Условное распределение:

$ p_{\xi|\eta=y_0} (x) = \cfrac{p_{\xi, \eta}  (x, y_0)}{p_\eta  (y_0)} $

Нормальное распределение:

$ \xi \sim \mathcal{N}(a, \sigma^2) $

$ p_\xi (x) = \cfrac{1}{\sigma\sqrt{2\pi}} \cdot \exp \left(-\cfrac{1}{2}\left(\cfrac{x-a}{\sigma}\right)^2\right) $

## Номер 1
Плотность двумерного нормального распределения имеет вид:
$p_{{\xi},{\eta}}(x,y)=C\cdot \exp{(-\frac{1}{2}(3x^{2}-3xy + 7y^{2} +15x +5y +25))}$

Задание 1.
Вычислить вектор математических ожиданий и ковариационные характеристиики данного случайного вектора. Построить график плотности распределния $p_{{\xi},{\eta}}$

Исходное выражение преобразуем(выражение из скобок):
$3(x^2-xy+\frac{1}{4}y^2)+\frac{25}{4}y^2 + 15x + 5y+25 = 3(x-\frac{1}{2}y)^2 + \frac{25y^2}{4} = 15(x-\frac{1}{2}y) + \frac{25}{4}y + 25 = 3((x-\frac{1}{2}y)^2 + 5(x-\frac{1}{2}y) + \frac{25}{4}) + \frac{25y^2}{4} + \frac{25}{2}y + \frac{25}{4} = 3((x-\frac{1}{2}y) + \frac{5}{2})^2 + \frac{25}{4}(y+1)^2 = 3((x+3) - \frac{1}{2}(y+1))^2 + \frac{25}{4}(y+1)^2 = 3(x+3)^2 - 3(x+3)(y+1) + 7(y+1)^2$

Вектор математических ожиданий:

$ \mathbb{E} \begin{pmatrix}
\xi\\
\eta
\end{pmatrix} = \begin{pmatrix}
-3\\
-1
\end{pmatrix} $

Обратная ковариационная матрица:

$ \Sigma^{-1} = \begin{pmatrix}
3 & \ -3/2\\
-3/2 & 7
\end{pmatrix} $

Обратим ковариационную матрицу:

$ \Sigma = \cfrac{4}{75}\begin{pmatrix}
7 & \ 3/2\\
3/2 & 3
\end{pmatrix} $

Константа С:

$C = \cfrac{1}{\left(2\pi\right)^{n/2} \sqrt{\det\Sigma}} = \cfrac{5\sqrt{3}}{4\pi}$

Итог:

$ p_{\xi, \eta}  (x, y) = \cfrac{5\sqrt{3}}{4\pi} \cdot \exp \left(-\cfrac{1}{2}\left(3x^2-3xy+7y^2+15x+5y+25\right)\right) $


```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.arange(-5, -1, 0.1) 
y = np.arange(-3, 1, 0.1) 
xgrid, ygrid = np.meshgrid(x, y) 
zgrid = 5 * 3 ** 0.5 / (4 * 3.14) * 2.72 ** (-0.5 * (3 * xgrid ** 2 - 3 * xgrid * ygrid + 7 * ygrid ** 2  + 15* xgrid + 5 * ygrid + 25))
fig = plt.figure() 
ax = plt.axes(projection="3d") 
ax.plot_surface(xgrid, ygrid, zgrid) 
plt.show()
```
