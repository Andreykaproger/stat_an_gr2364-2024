## Выполнение работы

Плотность двумерного нормального распределения имеет вид:
$p_{\xi,\eta}(x, y) = C \cdot \exp\left(-\frac{1}{2} \cdot (5x^2 - 4xy + 8y^2 - 6x - 12y + 9)\right);$

###1. Вычислим вектор математических ождианий и ковариационные характеристики данного случайного вектора, для этого сначала приведем квадратичную форму $Q(x, y)$ к каноническому виду:
$Q(x, y) =5x^2 - 4xy + 8y^2 - 6x - 12y + 9 = 5(x^2 - 2x + 1) - 4xy + 8y^2 - 4x - 12y + 4 = 5(x^2 - 2x + 1) + 8(y^2 - 2y + 1) - 4xy + 4x + 4y - 4 = 5(x^2 - 2x + 1) + 8(y^2 - 2y + 1) - 4(xy - x - y + 1) = 5(x^2 - 2x + 1) + 8(y^2 - 2y + 1) + 4\left(y(x + 1) - (x +-1)\right) = 5(x - 1)^2 - 4(x - 1)(y - 1) + 8(y - 1)^2$\
В результате приведения к каноническому виду квадратичной формы получаем:\
$p_{\xi,\eta}(x, y) = C \cdot \exp\left(-\frac{1}{2} \cdot \left[5(x - 1)^2 + 4(x - 1)(y - 1) + 8(y - 1)^2\right]\right).$\
1) Вектор математических ожиданий данного случайного вектора имеет вид:\
$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 1 \\ 1 \end{pmatrix}.$\
2) Матрица $\Sigma^{-1}:$\
$\Sigma^{-1} = \begin{pmatrix} 5 & 2 \\ 2 & 8 \end{pmatrix}.$\
3) Ковариационная матрица $\Sigma:$\
$\Sigma = \frac{1}{|\Sigma^{-1}|} \cdot \begin{pmatrix} 8 & -2 \\ -2 & 5 \end{pmatrix} = \frac{1}{36} \cdot \begin{pmatrix} 8 & -2 \\ -2 & 5 \end{pmatrix}.$\
$\Sigma = \begin{pmatrix} \mathbb{D}\xi & \mathop{\mathrm{cov}}\limits(\xi, \eta) \\ \mathop{\mathrm{cov}}\limits(\eta, \xi) & \mathbb{D}\eta \end{pmatrix},$ следовательно\
$\mathbb{D}\xi = \frac{1}{36} \cdot 8 = \frac{8}{36};$\
$\mathbb{D}\eta = \frac{1}{36} \cdot 5 = \frac{5}{36};$\
$\mathop{\mathrm{cov}}\limits(\xi, \eta) = \mathop{\mathrm{cov}}\limits(\eta, \xi) = \frac{1}{36} \cdot -2 = -\frac{2}{36} = -\frac{1}{18};$\
${\mathop{\mathrm{\rho}}\limits(\xi, \eta)} = \frac{\mathop{\mathrm{cov}}\limits(\xi, \eta)}{\sqrt{\mathbb{D}\xi\cdot\mathbb{D}\eta}} = \frac{-\frac{2}{36}}{\sqrt{\frac{8}{36} \cdot \frac{5}{36}}} = -\frac{\sqrt{10}}{10};$\
$C = \frac{1}{2\pi \cdot \sqrt{|\Sigma|}} = \frac{1}{2\pi \cdot \sqrt{\frac{1}{36}}} = \frac{1}{\frac{\pi}{3}}.$


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def p(x, y):
    coefficient = np.sqrt(36) / (2 * np.pi)
    exponent = -0.5 * (5 * x**2 - 4 * x * y + 8 * y**2 - 6 * x - 12 * y + 9)
    return coefficient * np.exp(exponent)

x_values = np.linspace(-4, 6, 100)
y_values = np.linspace(-4, 6, 100)

X, Y = np.meshgrid(x_values, y_values)

Z = p(X, Y)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title(r'$p_{\xi, \eta}(x, y) = \frac{\sqrt{36}}{2\pi} \cdot \exp \left(-\frac{1}{2}(5x^2 - 4xy + 8y^2 - 6x - 12y + 9)\right)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('p(x, y)')

plt.show()
```

### 3. Найдем ортогональное преобразование, переводящее соответствующий центрированный вектор в вектор с независимыми компонентами.
Найдем собственные числа и собственные вектора для матрицы $\Sigma^{-1}$:\
$\mathop{\mathrm{det}}(\Sigma^{-1} - \lambda \cdot E) = 0$\
$\left|\begin{pmatrix} 5 & 2 \\ 2 & 8 \end{pmatrix} - \begin{pmatrix} \lambda & 0 \\ 0 & \lambda \end{pmatrix}\right| = \left| \begin{pmatrix} 5 - \lambda & 2 \\ 2 & 8 - \lambda\end{pmatrix} \right| = 0$\
$(5 - \lambda)(8 - \lambda) - 4 = 0$\
$\lambda_{1} = 9; \lambda_{2} = 4$

Найдём собственные вектора:

$ \lambda_1 = 9: \begin{pmatrix}5-9 & 2 \\ 2&8-9\end{pmatrix} = \begin{pmatrix}-4 & 2 \\ 2&-1\end{pmatrix} $

$ \lambda_2 = 4: \begin{pmatrix}5-4 & 2 \\ 2&8-4\end{pmatrix} = \begin{pmatrix}1 & 2 \\ 2&4\end{pmatrix} $

$ v_2 = \alpha \begin{pmatrix}

Матрицу $ Q^{\intercal} $:

$ Q^{\intercal} = \cfrac{1}{\sqrt{5}} \begin{pmatrix}

\end{pmatrix} \Rightarrow Q = \cfrac{1}{\sqrt{5}} \begin{pmatrix}

$ \Sigma_{id} = Q \Sigma Q^{\intercal} = \cfrac{1}{36 \cdot 5} \begin{pmatrix} 4&8 \\ 2 & -1 \end{pmatrix} \begin{pmatrix} 8&-2 \\ -2 & 5 \end{pmatrix} \begin{pmatrix} 4&8 \\ 2 & -1 \end{pmatrix} = \cfrac{1}{36 \cdot 5} \begin{pmatrix} 20&0 \\ 0 & 45 \end{pmatrix} = \begin{pmatrix} 1/9&0 \\ 0 & 1/4 \end{pmatrix} $

$ \Sigma_{id}^{-1} = \begin{pmatrix} 9&0 \\ 0 & 4 \end{pmatrix} $

###4. Вычислим характеристики совместного распределения случайного вектора $(2\xi + 4\eta, 4\xi + 2\eta)^{\intercal}$ и записать его плотность.

$ \begin{pmatrix}2\xi + 4\eta \\ 4\xi + 2\eta\end{pmatrix} = \begin{pmatrix}2 & 4 \\ 4 & 2\end{pmatrix}\begin{pmatrix}\xi \\ \eta\end{pmatrix} $

$ \mathbb{E} \begin{pmatrix}2\xi + 4\eta \\ 4\xi + 2\eta\end{pmatrix} = \begin{pmatrix}2 & 4 \\ 4 & 2\end{pmatrix} \left(\mathbb{E} \begin{pmatrix}\xi\\\eta\end{pmatrix}\right) = \begin{pmatrix}2 & 4 \\ 4 & 2\end{pmatrix} \begin{pmatrix}1\\1 \end{pmatrix} = \begin{pmatrix}2 + 4 \\ 4 + 2 \end{pmatrix} = \begin{pmatrix}6 \\ 6\end{pmatrix}$

Пусть $ B = \begin{pmatrix}2 & 4 \\ 4 & 2\end{pmatrix} $

Тогда:

$ \Sigma_{2} = \mathrm{Var} \begin{pmatrix}2\xi + 4\eta \\ 4\xi + 2\eta\end{pmatrix} = B \Sigma B^{\intercal} = \cfrac{1}{36} \begin{pmatrix}2 & 4 \\ 4 & 2\end{pmatrix} \begin{pmatrix}8 & -2 \\ -2 & 5\end{pmatrix} \begin{pmatrix}2 & 4 \\ 4 & 2\end{pmatrix} = \cfrac{1}{36} \begin{pmatrix}80 & 64 \\ 64 & 116\end{pmatrix} = \cfrac{1}{9} \begin{pmatrix}20 & 16 \\ 16 & 29\end{pmatrix} $
\
\
$ \mathbb{D} \xi_2 = \cfrac{20}{9} $

$ \mathbb{D} \eta_2 = \cfrac{29}{9} $
\
\
$ \mathrm{cov} (\xi_2, \eta_2) = \cfrac{16}{9} $

$ \rho (\xi_2, \eta_2) = \cfrac{\mathrm{cov} (\xi_2, \eta_2)}{\sqrt{\mathbb{D} \xi_2 \cdot \mathbb{D} \eta_2}} = \frac{8\sqrt{145}}{145} $
\
\
$ \det \Sigma_{2} = 4 $

$ \Sigma_{2}^{-1} = \cfrac{1}{36} \begin{pmatrix}29 & -16 \\ -16 & 20\end{pmatrix} $

$ p_{\xi_2, \eta_2}  (x, y) = \cfrac{1}{2\pi \sqrt{4}} \cdot \exp \left(-\cfrac{1}{2}\cdot\cfrac{1}{36}\left(x-6, y-6\right)\begin{pmatrix}
29 & -16\\
-16 & 20
\end{pmatrix}\begin{pmatrix}
x-6\\
y-6
\end{pmatrix}\right) = \cfrac{1}{4\pi} \cdot \exp \left(-\cfrac{1}{2}\cdot\cfrac{1}{36}\left(29\left(x-6\right)^2 - 32\left(x-6\right)\left(y-6\right) + 20\left(y-6\right)^2\right)\right) =$

$ = \cfrac{1}{4\pi} \cdot \exp \left(-\cfrac{1}{2}\left(\cfrac{29}{36}\left(x-6\right)^2 - \cfrac{8}{9}\left(x-6\right)\left(y-6\right) + \cfrac{5}{9}\left(y-6\right)^2\right)\right) $

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def p(x, y):
    coefficient = 1 / (4 * np.pi)
    exponent = -0.5 * (29/36 * (x - 6)**2 - (8/9) * (x - 6) * (y - 6) + (5/9) * (y - 6)**2)
    return coefficient * np.exp(exponent)

x_values = np.linspace(0, 12, 100)
y_values = np.linspace(0, 12, 100)

X, Y = np.meshgrid(x_values, y_values)

Z = p(X, Y)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title(r'${\xi_2, \eta_2}(x, y) = \frac{1}{4\pi} \cdot \exp\left(-\frac{1}{2}\left(\frac{29}{36}(x-6)^2 - \frac{8}{9}(x-6)(y-6) + \frac{5}{9}(y-6)^2\right)\right)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('p(x, y)')

plt.show()
```

### 5. Найти условное распределение $ \xi $ при условии $ \eta $. Вычислить $ \mathbb{E} \left(\xi|\eta\right) $ и $ \mathbb{D} \left(\xi|\eta\right) $.
