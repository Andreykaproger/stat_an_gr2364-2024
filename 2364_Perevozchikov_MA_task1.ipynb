## Основные теоретические положения

**Плотность нормального распределения:**

$p_{xi} (x) = \cfrac{1}{\sqrt{2\pi\sigma^2}}\exp\left( \cfrac{-(x-a)^2}{2\sigma^2} \right)$

**Многомерное нормальное распределение**

Определение: Случайный вектор $\vec{\xi} = (\xi_1,\ldots,\xi_n)^{\intercal}:\Omega\rightarrow\mathbb R^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\vec{\zeta}=(\zeta_1,\ldots,\zeta_m)^{\intercal}$ вещественный вектор $\vec{\mu}=(\mu_1,\ldots,\mu_n) $ и ненулевая матрица $A\;\;\;n\times m$ размера такая, что $\xi=A\vec{\zeta}+\mu$,

где $\mu$ - вектор мат ожиданий, $\sigma_{i,j}=cov(\xi_i, \xi_j)$ - матрица ковариаций, $\Sigma=(\sigma_{i,j})_{1\leq i;\;j\leq n}$

**Плотность случайного вектора:**

$p_{\vec{\xi}}=(x_1,\ldots,x_n)=\cfrac{1}{(2\pi)^{n/2}\cdot\sqrt{\det \Sigma}}\exp\left(-\frac{1}{2}(\vec{x}-\mu)^{\intercal}\Sigma^{-1}(\vec{x}-\mu)\right)$

**Независимое распределение**

$\Sigma_{id}=Q\Sigma Q^{\intercal}=
    \begin{pmatrix}
        \frac{1}{\lambda_1}\; 0  \\
        0 \; \frac{1}{\lambda_2}
    \end{pmatrix}
$

**Условное распределение**

$p_{\xi | \eta}(x,y)=\cfrac{p_{\xi,\eta}(x,y)}{p_{\eta}(y)}\;\;\;\;$, где $p_{\eta}(y)==\cfrac{1}{\sqrt{2\pi\sigma^2}}\exp\left(-\cfrac{(y-\mu)^2}{2\sigma^2}\right)$

**Аффинное преобразование**

Это такое преобразование, которое переводит вектор в стандартный нормальный.

$
    \begin{pmatrix}
        u  \\
        v
    \end{pmatrix}
    \rightarrow
    \begin{pmatrix}
        x \\
        y
    \end{pmatrix}
$  ,

где
$
    \begin{pmatrix}
        x \\
        y
    \end{pmatrix}
    =
    M\cdot
    \begin{pmatrix}
        u \\
        v
    \end{pmatrix}
    +
    \begin{pmatrix}
        \alpha \\
        \beta
    \end{pmatrix}
$   ,

где $M$ - квадратичная матрица размером как вектор (в данном случае $2\times2$), $
    \begin{pmatrix}
        \alpha \\
        \beta
    \end{pmatrix}$
     - смещение

### Задание 1

Преобразуем полином из начала:

$3x^2-3xy+7y^2+9x-17y+13=$

$=3\left(x^2-xy+\cfrac{1}{4}y^2\right)-\cfrac{3}{4}y^2+7y^2+9x-17y+13=$

$=(3\left(x-\cfrac{1}{2}y\right)^2+3\cdot2\cdot\cfrac{3}{2}\left(x-\cfrac{1}{2}y+3\cdot\cfrac{9}{4}\right)-9\left(x-\cfrac{1}{2}y\right)-\cfrac{27}{4}+\cfrac{25}{4}y^2+9x-17y+13=$

$=3\left(x-\cfrac{1}{2}y+\cfrac{3}{2}\right)^2+\cfrac{25}{4}(y^2-2y+1)=$ 

$=3\left( (x+1)-\cfrac{1}{2}(y-1) \right)^2+\cfrac{25}{4}(y-1)^2=$

$=3(x+1)^2-3(x+1)(y-1)+7(y-1)^2$

Отсюда получаем вектор математического ожидания:

$\mathbb E \left( \xi; \eta \right)^{\intercal}=(-1;1)^{\intercal}$

Получим обратную матрицу:

$
\Sigma^{-1}=
    \begin{pmatrix}
        3\; -\frac{3}{2}  \\
        -\frac{3}{2} \; 7
    \end{pmatrix}
$

$ \det(\Sigma^{-1}) = \frac{75}{4}$

$
\Sigma=\frac{4}{75}\cdot
    \begin{pmatrix}
        7\; \frac{3}{2}  \\
        \frac{3}{2} \; 3
    \end{pmatrix}
$

$ \mathbb D_{\xi}=\frac{28}{75};\;\;\; \mathbb D_{\eta}=\frac{12}{75}$

$ cov(\xi,\eta)=\frac{12}{75} $

$corr = \frac{cov(\xi,\eta)}{\sqrt{\mathbb D_{\xi}\cdot \mathbb D_{\eta}}}=\sqrt{\frac{3}{7}}$

Вычислим значение $C$:

$C=\frac{1}{2\pi^{\frac{n}{2}}\cdot\sqrt{\det \Sigma}} = \frac{1}{2\pi \cdot\sqrt{\frac{4}{75}}} = \frac{5\sqrt{3}}{4\pi} $

Отсюда получаем, что функция распределения имеет вид:

$p_{\xi,\eta} (x,y) = \frac{5\sqrt{3}}{4\pi}\cdot\exp\left(-\cfrac{1}{2}(3x^2-3xy+7y^2+9x-17y+13)\right) $

$\;$

```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from math import pi, e

def func(x,y):
    return 5*3**0.5/(4*pi)*e**(-0.5*(3*x**2-3*x*y+7*y**2+9*x-17*y+13))

x=np.linspace(-3,3,50)
y=np.linspace(-3,3,50)
X, Y = np.meshgrid(x, y)
Z = func(X,Y)

fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot_wireframe(X,Y,Z, color ='green')
ax.set_title('Graph 1');
plt.show()

#https://www.geeksforgeeks.org/three-dimensional-plotting-in-python-using-matplotlib/
```
